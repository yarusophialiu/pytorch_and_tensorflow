## Series

```python
import numpy as np
import pandas as pd
```

```python
mylist = [1, 'a', 4.4]
pd.Series(data=mylist)
```

```python
d = {'a': 10, 'b': 20, 'c': 30}
pd.Series(data=d)
```

```python
labels = ['a', 'b', 'c']
pd.Series(data=mylist, index=labels)
```

## DataFrame



*   表格形数据结构，类似 sql





```python
np.random.seed(101)
rand_mat = np.random.randn(5, 4)
rand_mat
```

```python
df = pd.DataFrame(data=rand_mat)
df
```

```python
'A B C D'.split()
```

```python
df = pd.DataFrame(data=rand_mat, index='A B C D E'.split(), columns='W X Y Z'.split())
df
```

```python
## grabbing the column
df['W'], type(df['W'])
```

```python
mylist = ['W', 'Y']
df[['W', 'Y']]
```

```python
# create a new column
df['NEW'] = df['W'] + df['Y']
df
```

```python
# remove a column

df.drop('NEW', axis=1) # not permanently removed
```

```python
print(df)
df.drop('NEW', axis=1, inplace=True)
df
```

```python
# get the row
df
```

```python
df.loc['A']
print(df.loc[['A', 'E']])
```

```python
df.iloc[0]
print(df.iloc[[0, 3]])
```

```python
df.loc[['A', 'E']][['Y', 'Z']]
```

# group by

```python
data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
       'Sales':[200,120,340,124,243,350]}
df = pd.DataFrame(data)
df
```

```python
by_comp = df.groupby('Company')
by_comp
```

```python
by_comp.describe().transpose()['GOOG']
```

```python
df.columns, df.index, df.dtypes
```
