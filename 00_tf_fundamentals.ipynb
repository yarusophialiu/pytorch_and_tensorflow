# aggregation + cast


```python
print(A)
A = tf.cast(A, dtype=tf.float16)
print(A)
```

```python
# get absolute values
M = tf.constant([[-1, 3]]);
tf.abs(M)

# create a random tensor with values between 10, 100 of size 50
B = tf.constant(np.random.randint(0, 100, size=50))
B

# find the max, min, mean, sum 
tf.reduce_max(B), tf.reduce_min(B), tf.reduce_mean(B)

tf.size(B), B.shape, B.ndim
print(tf.reduce_min(B), np.min(B))
```

```python
# find variance, standard deviation
import tensorflow_probability as tfp;
print(A)
var = tfp.stats.variance(A)
std = tf.math.reduce_std(tf.cast(A, dtype=tf.float32))
var2 = tf.math.reduce_variance(tf.cast(A, dtype=tf.float32))
var, var2,std
```

## find argmin, argmax

```python
tf.random.set_seed(42)
F = tf.random.uniform(shape=[50])
F
```

```python
tf.argmax(F) # index
assert F[tf.argmax(F)] == tf.reduce_max(F)

print(A)
tf.argmax(A, axis=1)
```

squeezing a tensor (removing all single dim)


```python
tf.random.set_seed(42)
G = tf.constant(tf.random.uniform(shape=[16]), shape=(1, 1, 1, 16))
print(G, "\n")
G_squeezed = tf.squeeze(G)
print(G_squeezed, "\n", G_squeezed.shape)
```

# one hot encoding tensors

```python
some_list = [0, 1, 2, 1]
tf.one_hot(some_list, depth=4, on_value=999, off_value=0)
```

# square, squaer root, log

```python
H = tf.range(1, 10)
square = tf.square(H)
sqrt = tf.sqrt(tf.cast(H, dtype=tf.float32))
log = tf.math.log(tf.cast(H, dtype=tf.float16))
square, sqrt, log
```
