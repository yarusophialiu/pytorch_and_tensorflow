## Indexing

```python
arr = np.arange(0, 10)
arr
```

```python
# point to original array
slice_of_arr = arr[:5]
slice_of_arr
```

```python
slice_of_arr[:] = 99
slice_of_arr, arr
```

```python
# create a copy in the memory
arr_copy = arr.copy()
arr_copy[:] = 100
arr_copy, arr
```

```python
arr_2d = np.arange(5, 50, 5)
arr_2d = arr_2d.reshape(3, 3)
arr_2d
```

```python
# get subset of the matrix
arr_2d[:2, 1:]
```
